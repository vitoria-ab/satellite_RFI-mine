```python
import pickle
import sys
import time
from datetime import datetime
import pytz
```

```python
"""
User enters the date and time of observation with respect to UTC time

format: YYYY, MM, DD, hh, mm, ss
"""

obs_time = datetime(2019, 2, 25, 0, 40, 11)

fname = int((obs_time - datetime(1970, 1, 1)).total_seconds())

print "File name to be used is: ", fname
```

```python
print """
nd_s -         the scanning timestamps (can leave empty)

nd_s0 -        a list of the scan in seconds should be given here,
               this is added to the fname/obs_time and can be the 
               scanning zone, in seconds or timestamp seconds

freqs -        a list of the frequency for the observation 
               (f0, f1, f2....., fn)

timestamps -   can leave empty

nd_s0_coords - a tuple of (RA_list, Declination_list), these should be of 
               equal size and equal to the length of nd_s0

"""
```

```python
nd_s0 = None
nd_s0_pos = None
nd_s = None
nd_s0_coords = None
freqs = None
timestamps = None
```

```python
sats_info = {"nd_s0":nd_s0, 
             "nd_s0_pos":nd_s0_pos,
             "nd_s":nd_s,
             "nd_s0_coords":nd_s0_coords,
             "frequency":freqs,
             "time":timestamps    
}
```

```python
# Where to save the file
data_save = ''

pickle.dump(sats_info, open(data_save+str(fname)+'_katdal_info.p', 'wb'))
```
